# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# your code goes here
QUERY = """
SELECT ConvertedComp FROM master

"""

df = pd.read_sql_query(QUERY, conn)

plt.rcParams["figure.figsize"] = [12, 8]

ax = sns.histplot(data = df["ConvertedComp"], bins = 150)
ax.set_title("Histogram: ConvertedComp", size = 20)
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age FROM master

"""

df = pd.read_sql_query(QUERY, conn)

plt.rcParams["figure.figsize"] = [12, 8]

ax = sns.boxplot(data = df["Age"], orient = "h")
ax.set_title("Box Plot: Age", size = 20)
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs FROM master

"""

df = pd.read_sql_query(QUERY, conn)

plt.rcParams["figure.figsize"] = [12, 8]
ax = sns.scatterplot(data = df, x = "Age", y = "WorkWeekHrs")
ax.set_title("Scatter Plot: Age V/s WorkWeekHrs", size = 20)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master

"""

df = pd.read_sql_query(QUERY, conn)

plt.rcParams["figure.figsize"] = [12, 8]

ax = sns.scatterplot(data = df, x = "CodeRevHrs", y = "WorkWeekHrs", size = "Age", hue = "Age", alpha = 0.6, sizes=(1, 1000), palette="Spectral")
ax.set_title("Bubble Plot: CodeRevHrs V/s WorkWeekHrs with Age for Bubble Size", size = 20)
```

### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT *, COUNT(*) AS Count FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5

"""

df = pd.read_sql_query(QUERY, conn)

data = df["Count"]
labels = df["DatabaseDesireNextYear"]

plt.rcParams["figure.figsize"] = [10, 10]

fig, ax = plt.subplots()
ax.pie(data, labels = labels, autopct = "%1.2f%%", shadow = True, explode = (0, 0.1, 0, 0, 0))
plt.show()
```

```python
#In the list of most popular languages respondents wish to learn next year, what is the rank of Python?
QUERY = """
SELECT *, COUNT(*) AS Count FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY Count DESC
LIMIT 5

"""

df = pd.read_sql_query(QUERY, conn)

data = df["Count"]
labels = df["LanguageDesireNextYear"]

plt.rcParams["figure.figsize"] = [10, 10]

fig, ax = plt.subplots()
ax.pie(data, labels = labels, autopct = "%1.2f%%", shadow = True, explode = (0, 0, 0.1, 0, 0))
plt.show()
```

```python
#How many respondents indicated that they currently work with 'SQL'? 
QUERY = """
SELECT *, COUNT(*) AS Count FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY Count DESC
LIMIT 5

"""

df = pd.read_sql_query(QUERY, conn)
print(df)

data = df["Count"]
labels = df["LanguageWorkedWith"]

plt.rcParams["figure.figsize"] = [10, 10]

fig, ax = plt.subplots()
ax.pie(data, labels = labels, autopct = "%1.2f%%", shadow = True, explode = (0, 0, 0.1, 0, 0))
plt.show()
```

```python
#How many respondents indicated that they work on 'MySQL' only? 
QUERY = """
SELECT * FROM DatabaseWorkedWith
ORDER BY Respondent

"""

df = pd.read_sql_query(QUERY, conn)

df_mysql = df.groupby(["Respondent"]).sum()
mysql = df_mysql[df_mysql["DatabaseWorkedWith"] == "MySQL"].count()
print("MySQL", mysql)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35

"""

df = pd.read_sql_query(QUERY, conn)

ww = df["WorkWeekHrs"].quantile(0.5)
cr = df["CodeRevHrs"].quantile(0.5)

plt.rcParams["figure.figsize"] = [12, 8]

fig, ax = plt.subplots()

ax.bar("Age 30 - 35", ww, label = "WorkWeekHrs")
ax.bar("Age 30 - 35", cr, bottom = ww, label = "CodeRevHrs")
ax.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age Between 45 AND 60

"""

df = pd.read_sql_query(QUERY, conn)

m = df.groupby("Age")["ConvertedComp"].median()

data = pd.DataFrame(list(zip(m.index, m.values)), columns = ["Age", "ConvertedComp_Median"])

plt.rcParams['figure.figsize'] = [12, 8]

ax = sns.lineplot(data = data, x = "Age", y = "ConvertedComp_Median")
ax.set_title("Line Chart: Median of ConvertedComp With Age", size = 20)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) AS Frequency FROM master
GROUP BY MainBranch

"""
df = pd.read_sql_query(QUERY, conn)

plt.rcParams["figure.figsize"] = [10, 6]

ax = sns.barplot(data = df, y = "MainBranch", x = "Frequency", orient = "h")
ax.set_title("Bar Chart: MainBranch", size = 20)
plt.show()
```

```python
#Based on the Line Chart of Age and median ConvertedComp for the age group 25-30, what kind of correlation do you see between Age and median ConvertedComp?
QUERY = """
SELECT Age, ConvertedComp FROM master
WHERE Age BETWEEN 25 AND 30

"""

df = pd.read_sql_query(QUERY, conn)

m = df.groupby("Age")["ConvertedComp"].median()
data = pd.DataFrame(list(zip(m.index, m.values)), columns = ["Age", "ConvertedComp_Median"])

plt.rcParams["figure.figsize"] = [12, 8]

sns.lineplot(data = data, x = "Age", y = "ConvertedComp_Median")
plt.show()
```
